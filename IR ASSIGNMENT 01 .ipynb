```python
import string
import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords as sws
def tokenize_text(html_content):
        stop_words = set(sws.words('english'))
        text_tokens = nltk.word_tokenize(html_content.translate(str.maketrans('', '', string.punctuation)))

        txt = []

        for i in text_tokens:
            if(i.lower() not in stop_words):
                txt.append(i)

        cont = [word for word in txt if word.strip()]
        s=' '.join(cont)
        return s
```
